```python
import numpy as np
import math
```

```python
def aimFun1(x):
    y = x**3 - 60*x**2 - 4*x + 6
    return y
```

```python
def x1(x):
    x1 = 3*x**2 - 120*x - 4
    return x1
```

```python
x = np.random.uniform(low=0,high=100) #随机一个起点
y_hat = aimFun1(x)

lr = 1e-3
iter_num = 0
X1 = [x]
Y1 = [y_hat]

while(iter_num<30):
    tmp_x = x - lr*x1(x)
    iter_num += 1
    y_hat = aimFun1(tmp_x)
    
    X1.append(tmp_x)
    Y1.append(y_hat)
    x = tmp_x
    

print(X1)
print(Y1)
```

```python
import matplotlib.pyplot as plt
plot_x=np.linspace(20,80,500)   #在-1到6之间等距的生成141个数
plot_y= (plot_x)**3 - 60*(plot_x)**2 - 4 *(plot_x) + 6	   # 同时根据plot_x来生成plot_y

def J(x):
     return x**3 - 60*x**2 - 4*x + 6

plt.plot(plot_x,plot_y)     
plt.plot(np.array(X1),J(np.array(X1)),color='r',marker='*')   #绘制x的轨迹
plt.show()
```

```python
def aimFun2(x):
    y = -x**2 + 1
    return y

def x2(x):
    x2 = -2*x
    return x2
```

```python
x = np.random.uniform(low=-4,high=4) #随机一个起点
y_hat = aimFun2(x)

lr2 = 0.005
iter_num = 0
X2 = [x]
Y2 = [y_hat]

while(iter_num<30):
    tmp_x = x - lr2*x2(x)
    iter_num += 1
    y_hat = aimFun2(tmp_x)
    
    X2.append(tmp_x)
    Y2.append(y_hat)
    x = tmp_x
    

print(X2)
print(Y2)
```

```python
import matplotlib.pyplot as plt
plot_x=np.linspace(-4,4,30)   #在-1到6之间等距的生成141个数
plot_y= -(plot_x)**2 + 1	   # 同时根据plot_x来生成plot_y

def J(x):
     return -x**2 + 1

plt.plot(plot_x,plot_y)     
plt.plot(np.array(X2),J(np.array(X2)),color='r',marker='*')   #绘制x的轨迹
plt.show()
```
